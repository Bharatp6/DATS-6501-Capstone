```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

### Correlation matrix function

```python
def correlation_mat(ticker_us, ticker_asian, ticker_asian_index, ticker_american_index):
  adr = yf.download(ticker_us)
  asian = yf.download(ticker_asian)
  index_asian = yf.download(ticker_asian_index)
  nasdaq = yf.download(ticker_american_index)

  adr = adr.reset_index()
  asian = asian.reset_index()
  index_asian = index_asian.reset_index()
  nasdaq = nasdaq.reset_index()

  market = 'adr'
  adr = adr.rename(columns = {'Date':'Date','Open':f'Open_{market}','High':f'High_{market}','Low':f'Low_{market}','Close':f'Close_{market}','Adj Close':f'Adj Close_{market}',"Volume":f'Volume_{market}'})

  market = 'asian'
  asian = asian.rename(columns = {'Date':'Date','Open':f'Open_{market}','High':f'High_{market}','Low':f'Low_{market}','Close':f'Close_{market}','Adj Close':f'Adj Close_{market}',"Volume":f'Volume_{market}'})

  market = 'asian_ind'
  index_asian = index_asian.rename(columns = {'Date':'Date','Open':f'Open_{market}','High':f'High_{market}','Low':f'Low_{market}','Close':f'Close_{market}','Adj Close':f'Adj Close_{market}',"Volume":f'Volume_{market}'})

  market = 'nasdaq'
  nasdaq = nasdaq.rename(columns = {'Date':'Date','Open':f'Open_{market}','High':f'High_{market}','Low':f'Low_{market}','Close':f'Close_{market}','Adj Close':f'Adj Close_{market}',"Volume":f'Volume_{market}'})


  merged = adr.merge(asian, on = 'Date', how = 'inner')
  merged = merged.merge(index_asian, on = 'Date', how = 'inner')
  merged = merged.merge(nasdaq, on = 'Date', how = 'inner')

  correlation_matrix = merged.corr()

  return correlation_matrix
```

### Correlation matrix for INFOSYS

```python
correlation_mat('INFY','INFY.NS','^BSESN','^IXIC')
```

### Correlation matrix for HDFC BANK

```python
correlation_mat('HDB','HDFCBANK.NS','^BSESN','^IXIC')
```

```python
correlation_mat('WIT','WIPRO.NS','^BSESN','^IXIC')
```
